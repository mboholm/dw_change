```python
import json
import pandas as pd
```

```python
def extok_stats(file):
    # "/home/max/Corpora/fb-pol-corpus/flashback-pol-time/counter.log"
    with open(file, "r") as f:
        counts = json.loads(f.read())
        
    df = pd.DataFrame(counts).transpose()
    
    print(df)
    print()
    print(df.plot())
```

```python
extok_stats("/home/max/Corpora/fb-pol-corpus/flashback-pol-time/counter.log")
```
