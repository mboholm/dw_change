```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
from difflib import SequenceMatcher
from pathlib import Path
from sklearn.utils.extmath import softmax
from collections import Counter
```

```python
def PairwiseMeanSimilarity(v, v_list):
    
    pairwise = cosine_similarity(v, v_list)
    pairwise_mean = pairwise.mean()
    
    return pairwise_mean
```

```python
import random
random.seed(545)
```

```python
INGROUPvec = np.random.rand(100,200)
OUTGROUPvec = np.random.rand(100,200)
```

```python
ING_centroid  = INGROUPvec.mean(axis=0)
OUTG_centroid = OUTGROUPvec.mean(axis=0)
```

```python
results = []
for year in range(11):
    vector = np.random.rand(1,200)

    i_cnt = cosine_similarity(vector.reshape(1,-1), ING_centroid.reshape(1,-1))[0][0] 
    o_cnt = cosine_similarity(vector.reshape(1,-1), OUTG_centroid.reshape(1,-1))[0][0]

    i_pwn = PairwiseMeanSimilarity(vector.reshape(1, -1), INGROUPvec)
    o_pwn = PairwiseMeanSimilarity(vector.reshape(1, -1), OUTGROUPvec)
    
    d = {
    "I-cnt": i_cnt,
    "O-cnt": o_cnt ,
    "cnt-ssc": i_cnt / (i_cnt + o_cnt),
    "cnt-smx": softmax([[i_cnt, o_cnt]])[0][0],

    "I-pwn": i_pwn,
    "O-pwn": o_pwn,
    "pwn-ssc": i_pwn / (i_pwn + o_pwn),
    "pwn-smx": softmax([[i_pwn, o_pwn]])[0][0]
    }
    results.append(d)
```

```python
df = pd.DataFrame(results, index=list(range(2000, 2011))).transpose()
```

```python
df.round(3)
```

```python
df.transpose().corr().round(3)
```

Although *X*-cnt and *x*-pwn are different, their correlation is almost perfect.
